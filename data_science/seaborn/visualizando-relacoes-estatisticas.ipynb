## Gráfico de Dispersão

O gráfico de dispersão é um dos pilares da visualização estatística. Ele descreve a distribuição conjunta de duas variáveis usando uma nuvem de pontos, onde cada ponto representa uma observação no conjunto de dados. Essa representação permite que o olho deduza uma quantidade substancial de informações sobre se existe algum relacionamento significativo entre elas.

Existem várias maneiras de desenhar um gráfico de dispersão no seaborn. O mais básico, que deve ser usado quando ambas as variáveis são numéricas, é a ``scatterplot()``.O ``scatterplot()``é o padrão kind em ``relplot()``(também pode ser forçado pela configuração kind="scatter"):

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", data=tips);
```

Enquanto os pontos são plotados em duas dimensões, outra dimensão pode ser adicionada à plotagem colorindo os pontos de acordo com uma terceira variável. Em ``seaborn``, isso é chamado de “semântica de matiz”, porque a cor do ponto ganha significado:

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

Para enfatizar a diferença entre as classes e melhorar a acessibilidade, você pode usar um estilo de marcador diferente para cada classe:

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker",
            data=tips);
```

Também é possível representar quatro variáveis alterando o matiz e o estilo de cada ponto de forma independente. Mas isso deve ser feito com cuidado, porque o olho é muito menos sensível à forma do que à cor:

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", data=tips);
```

Nos dois casos, você pode personalizar a paleta de cores. Existem muitas opções para o fazer. Aqui, nós personalizamos uma paleta seqüencial usando a interface de string para ``cubehelix_palette()``:

```python
sns.relplot(x="total_bill", y="tip", hue="size", palette="ch:r=-.5,l=.75", data=tips);
```

## Gráficos de Linhas

Os gráficos de dispersão são altamente eficazes, mas não existe um tipo universalmente ideal de visualização. Em vez disso, a representação visual deve ser adaptada para as especificidades do conjunto de dados e para a pergunta que você está tentando responder com a plotagem.

Com alguns conjuntos de dados, você pode querer entender as alterações em uma variável como uma função do tempo ou uma variável contínua similar. Nesta situação, uma boa escolha é desenhar um gráfico de linha. No ``seaborn``, isso pode ser feito pela ``lineplot()``, diretamente ou com a ``relplot()`` configuração kind="line":

```python
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()
```

Como ``lineplot()`` supõe que, na maioria das vezes, você está tentando desenhar y como uma função x, o comportamento padrão é classificar os dados pelos x valores antes de plotar. No entanto, isso pode ser desativado:

```python
df = pd.DataFrame(np.random.randn(500, 2).cumsum(axis=0), columns=["x", "y"])
sns.relplot(x="x", y="y", sort=False, kind="line", data=df);
```

Conjuntos de dados mais complexos terão várias medições para o mesmo valor da x variável. O comportamento padrão no seaborn é agregar as múltiplas medidas em cada x valor, traçando a média e o intervalo de confiança de 95% em torno da média:

```python
fmri = sns.load_dataset("fmri")
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri);
```

A ``lineplot()`` função tem a mesma flexibilidade de ``scatterplot()``: pode mostrar até três variáveis adicionais, modificando o matiz, o tamanho e o estilo dos elementos de plotagem. Ele faz isso usando a mesma API ``scatterplot()``, o que significa que não precisamos parar e pensar sobre os parâmetros que controlam a aparência de linhas versus pontos no matplotlib.

O uso da semântica ``lineplot()`` também determinará como os dados serão agregados. Por exemplo, adicionar uma semântica de matiz com dois níveis divide a plotagem em duas linhas e bandas de erro, colorindo cada uma delas para indicar qual subconjunto dos dados elas correspondem.

```python
sns.relplot(x="timepoint", y="signal", hue="event", kind="line", data=fmri);
```

Adicionar um estilo semântico a um gráfico de linha altera o padrão de traços na linha por padrão:

```python
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
            kind="line", data=fmri);
```

Mas você pode identificar subconjuntos pelos marcadores usados em cada observação, junto com os traços ou em vez deles:

```python
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
            dashes=False, markers=True, kind="line", data=fmri);
```

### Plotando com dados de data

Os gráficos de linha geralmente são usados para visualizar dados associados a datas e horas reais. Essas funções passam os dados para baixo em seu formato original para as funções ``matplotlib`` subjacentes e, assim, podem aproveitar a capacidade do ``matplotlib`` de formatar datas em rótulos de escala. Mas toda essa formatação terá que ocorrer na camada ``matplotlib``, e você deve consultar a documentação do ``matplotlib`` para ver como ela funciona:

```python
df = pd.DataFrame(dict(time=pd.date_range("2017-1-1", periods=500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()
```

## Mostrando vários relacionamentos de visualização

Quando você quer entender como um relacionamento entre duas variáveis depende de mais de uma outra variável?

A melhor abordagem pode ser fazer mais de um gráfico. Porque ``relplot()`` é baseado no ``FacetGrid``, isso é fácil de fazer. Para mostrar a influência de uma variável adicional, em vez de atribuí-la a um dos papéis semânticos na plotagem, use-a para “facetar” a visualização. Isso significa que você cria vários eixos e plota subconjuntos dos dados em cada um deles:

```python
sns.relplot(x="total_bill", y="tip", hue="smoker",
            col="time", data=tips);
```

Você também pode mostrar a influência de duas variáveis desta maneira: uma por lapidação nas colunas e uma por lapidação nas linhas. Quando você começar a adicionar mais variáveis à grade, poderá diminuir o tamanho da figura. Lembre-se de que o tamanho ``FacetGrid`` é parametrizado pela altura e proporção de cada faceta :

```python
sns.relplot(x="timepoint", y="signal", hue="subject",
            col="region", row="event", height=3,
            kind="line", estimator=None, data=fmri);
```

Quando você quiser examinar os efeitos em vários níveis de uma variável, pode ser uma boa ideia ajustar essa variável nas colunas e, em seguida, "envolver" as facetas nas linhas:

```python
sns.relplot(x="timepoint", y="signal", hue="event", style="event",
            col="subject", col_wrap=5,
            height=3, aspect=.75, linewidth=2.5,
            kind="line", data=fmri.query("region == 'frontal'"));
```
